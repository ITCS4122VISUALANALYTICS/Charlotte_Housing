```python
import streamlit as st
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('../data_sets/df_cleaned.csv')
df1 = pd.DataFrame(data)
df1['HINCP']=df1['HINCP']//12
df1["PUMA"].replace({3101 : "Charlotte City (Central)", 
    3102 : "Charlotte City (Northwest)", 
    3103 : "Charlotte City (Northeast)",
    3104 : "Charlotte City (South)",
    3105 : "Charlotte City (Southwest)", 
    3106 : "Mecklenburg County (North)",
    3107 : "Mecklenburg County (East)",
    3108 : "Mecklenburg County (South)"}, inplace=True)
df1.head(5)
```

```python

input_dropdown = alt.binding_select(options=['Charlotte City (Central)','Charlotte City (Northwest)','Charlotte City (Northeast)',
                                            'Charlotte City (South)', 'Charlotte City (Southwest)', 'Mecklenburg County (North)',
                                            'Mecklenburg County (East)', 'Mecklenburg County (South)'], name='PUMA: ')
selection = alt.selection_single(fields=['PUMA'], bind=input_dropdown)

alt.Chart(df1).mark_point(filled=True).encode(
    alt.X('RNTP', title = 'Monthly Rent'),
    alt.Y('HINCP', title = 'Monthly Household Income'),
    alt.Size('BDSP', title = 'Number of Bedrooms'),
    alt.Color('NP', title = 'Number of Persons'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('RNTP:Q', title = 'Monthly Gross Rent'),
               alt.Tooltip('HINCP:Q', title = 'Monthly Household Income'),
               alt.Tooltip('NP:Q', title = 'Number of Persons'),
               alt.Tooltip('BDSP:Q', title = 'Number of Bedrooms')
              ]
).add_selection(
    selection
).transform_filter(
    selection
)
```
